```python
import json
from multiprocessing import Pool
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import operator
from CSMrouge import RRRouge
import random
from collections import Counter
import itertools
from functools import reduce
```

```python
plt.rcParams.update({'font.size': 18})

def plot_experiment_stats():
    fopen = open('experiment_stats.json', 'r')
    data = json.load(fopen)
    training_loss = data["training_loss"]
    dev_loss = data["dev_loss"]

    training_rouge = data["training_rouge"]
    dev_rouge = data["dev_rouge"]

    ### Loss
    plt.figure()
    plt.plot(range(1, 21), training_loss, label="Training Loss")
    plt.plot(range(1, 21), dev_loss, label = "Dev Loss")
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    locator = matplotlib.ticker.MultipleLocator(2)
    plt.gca().xaxis.set_major_locator(locator)
    formatter = matplotlib.ticker.StrMethodFormatter("{x:.0f}")
    plt.gca().xaxis.set_major_formatter(formatter)
    
#     plt.show()
    plt.savefig('Epoch_loss')
    
    ### Rouge
    plt.figure()
    plt.plot(range(1,21), training_rouge, label="Training Rouge")
    plt.plot(range(1,21), dev_rouge, label = "Dev Rouge")
    plt.xlabel('Epoch')
    plt.ylabel('Rouge')
    plt.legend()
    locator = matplotlib.ticker.MultipleLocator(2)
    plt.gca().xaxis.set_major_locator(locator)
    formatter = matplotlib.ticker.StrMethodFormatter("{x:.0f}")
    plt.gca().xaxis.set_major_formatter(formatter)
#     plt.show()  
    plt.savefig('Epoch_rouge')
    
plot_experiment_stats()
```

```python
def plot_best_dev_output():
    fopen = open('Best_dev_output.json', 'r')
    data = json.load(fopen)
    predictions, ans1, ans2 = [], [], []
    for dataPoint in data:
        predictions.append(dataPoint['prediction'])
        ans1.append(dataPoint['answer1'])
        ans2.append(dataPoint['answer2'])

    predictions_len = list(map(lambda x: len(x.split()), predictions))
    ans1_len = list(map(lambda x: len(x.split()), ans1))
    ans2_len = list(map(lambda x: len(x.split()), ans2))
    print("Mean len of predictions: %f" %(np.mean(predictions_len)))
    print("Mean len of ans1: %f" %(np.mean(ans1_len)))
    print("Mean len of ans2: %f" %(np.mean(ans2_len)))
    
plot_best_dev_output()
```

```python
def load_preprocess_data():
    rrrouge = RRRouge()    
    train_rouge_ans_choosen = []
    
    fopen = open('../prepro/narrativeqa_summary_test_index.json', 'r')
    data = json.load(fopen)
    context, question, answers, chosen_answer = [], [], [], []
#     print(data[0]['context']) # String
#     print(data[0]['question']) #S tring
#     print(data[0]['answers']) # List of String
#     print(data[0]['chosen_answer']) # List of String
    
    for dataPoint in data:
        idx = dataPoint['_id']
        c = dataPoint['context']
        q = dataPoint['question']
        a = dataPoint['answers']
        cho_ans = " ".join(dataPoint['chosen_answer'])
        r = rrrouge.calc_score([cho_ans], [a[0], a[1]]) #chosen answer with answers 1 and 2
        train_rouge_ans_choosen.append((idx +" " + a[0].lower() +" "+a[1].lower(), q, r))        
        
    return train_rouge_ans_choosen
train_rouge_ans_choosen = load_preprocess_data() ## Rouge between choosen answers and the answers1 and 2
```

```python
def get_train_rouge_ans_pred():

    fopen = open('Best_eval_output_corrected.json', 'r')
    data = json.load(fopen)
    train_rouge_ans_pred = []
    
    for dataPoint in data:
        idx = dataPoint['_id']
        p = dataPoint['prediction']
        ans1 = dataPoint['answer1']
        ans2 = dataPoint['answer2']
        r = dataPoint['rouge_score']
        train_rouge_ans_pred.append((idx + " " + ans1.lower() + " " + ans2.lower(), p, r)) ## rouge between the predicted answer and the ans 1 and 2
    return train_rouge_ans_pred

train_rouge_ans_pred = get_train_rouge_ans_pred()
```
